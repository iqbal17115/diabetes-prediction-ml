```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib

# ================================
# Function for reusable preprocessing
# ================================
def load_and_preprocess(raw_path, processed_path=None, scaler_path=None):
    """
    Load raw CSV data, handle missing values, scale features, optionally save processed data and scaler.
    
    Parameters:
        raw_path (str): Path to raw CSV file
        processed_path (str, optional): Path to save processed CSV
        scaler_path (str, optional): Path to save fitted scaler
    
    Returns:
        df_processed (pd.DataFrame): Preprocessed dataframe
    """
    
    # 1️⃣ Load raw data
    df = pd.read_csv(raw_path)
    
    # 2️⃣ Handle missing values
    df.fillna(df.median(), inplace=True)  # numeric columns
    # categorical handling if needed
    # df['CategoryColumn'].fillna(df['CategoryColumn'].mode()[0], inplace=True)
    
    # 3️⃣ Separate features and target
    features = df.drop("Outcome", axis=1)
    target = df["Outcome"]
    
    # 4️⃣ Scale features
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    # 5️⃣ Save scaler if path provided
    if scaler_path:
        joblib.dump(scaler, scaler_path)
        print(f"Scaler saved to {scaler_path}")
    
    # 6️⃣ Create processed dataframe
    df_processed = pd.DataFrame(features_scaled, columns=features.columns)
    df_processed['Outcome'] = target
    
    # 7️⃣ Save processed data if path provided
    if processed_path:
        df_processed.to_csv(processed_path, index=False)
        print(f"Processed data saved to {processed_path}")
    
    return df_processed

# ================================
# Optional: run as main script
# ================================
if __name__ == "__main__":
    raw_path = "../data/raw/diabetes.csv"
    processed_path = "../data/processed/diabetes_processed.csv"
    scaler_path = "../models/scaler.pkl"
    
    df = load_and_preprocess(raw_path, processed_path, scaler_path)
    print("Preprocessing complete.")
```
